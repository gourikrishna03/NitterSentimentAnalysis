This notebook demonstrates how sentiment analysis was applied to cleaned tweets.

It performs:
- loading cleaned tweet sample
- applying VADER/TextBlob sentiment
- classifying sentiment (pos/neu/neg)
- displaying results as a table

This is only to show the pipeline, not to re-run the full project.


```python
import pandas as pd

df = pd.read_csv("../data/processed/cleaned_sample.csv")
df.head()
```

### Apply Sentiment Analysis (example)
Using VADER sentiment analyzer.


```python
from nltk.sentiment import SentimentIntensityAnalyzer

sia = SentimentIntensityAnalyzer()

df["sentiment_score"] = df["cleaned_text"].apply(lambda x: sia.polarity_scores(str(x))["compound"])

def classify(score):
    if score >= 0.05:
        return "positive"
    elif score <= -0.05:
        return "negative"
    else:
        return "neutral"

df["sentiment"] = df["sentiment_score"].apply(classify)

df.head()
```

### Save demo sentiment output (optional)


```python
df.to_csv("../data/processed/sentiment_sample.csv", index=False)
```
