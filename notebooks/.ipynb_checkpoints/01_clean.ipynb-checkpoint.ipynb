## Objectives
- Load raw tweets
- Apply sample cleaning functions
- Save or preview cleaned outputs


```python
import pandas as pd

df = pd.read_csv("../data/raw/ilyasut_tweets.csv")
df.head()
```

### Example: Cleaning functions (demonstration only)


```python
import re

def clean_text(text):
    text = re.sub(r"http\S+", "", text)      # remove URLs
    text = re.sub(r"@\w+", "", text)         # remove mentions
    text = re.sub(r"[^A-Za-z0-9\s]", "", text)  # remove emojis/punctuation
    return text.strip()

df["cleaned_text"] = df["text"].astype(str).apply(clean_text)
df.head()
```

### Save cleaned sample (optional)


```python
df.to_csv("../data/processed/cleaned_sample.csv", index=False)
```
